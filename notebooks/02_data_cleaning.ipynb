# Day 3: Data Cleaning

## Objective

The goal of this notebook is to clean the Amazon Sales dataset by:
- Removing unnecessary columns
- Investigating missing values
- Removing duplicate records
- Converting data types
- Preparing the dataset for exploratory data analysis (EDA)

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/Amazon Sale Report.csv')
```

```python
df.shape
```

```python
df_clean = df.copy()
```

## Creating a Working Copy

A copy of the original dataset is created so that the raw dataset remains unchanged throughout the cleaning process.

```python
df_clean[['New', 'PendingS']].info()
```

```python
df_clean.drop(columns=['New', 'PendingS'], inplace=True)
```

```python
df_clean.shape
```

## Removing Empty Columns

The columns `New` and `PendingS` contain 100% missing values and do not provide any useful information for analysis.

These columns have been removed.

```python
df_clean.duplicated().sum()
```

```python
df_clean.drop_duplicates(inplace=True)
```

```python
df_clean.duplicated().sum()
```

## Removing Duplicate Records

Duplicate records can distort analysis results.

A total of 168 duplicate rows were identified and removed from the dataset.

```python
df_clean['Date'].dtype
```

```python
df_clean['Date'] = pd.to_datetime(df_clean['Date'])
```

```python
df_clean['Date'].dtype
```

## Converting Date Datatype

The Date column was originally stored as an object datatype.

It has been converted to datetime format to enable time-based analysis such as monthly sales trends and seasonal analysis.

```python
df_clean[df_clean['Amount'].isnull()]['Status'].value_counts()
```

```python
df_clean[df_clean['currency'].isnull()]['Status'].value_counts()
```

```python
df_clean['fulfilled-by'].value_counts(dropna=False)
```

```python
df_clean[['ship-city',
          'ship-state',
          'ship-postal-code',
          'ship-country']].isnull().sum()
```

```python
df_clean.dropna(
    subset=[
        'ship-city',
        'ship-state',
        'ship-postal-code',
        'ship-country'
    ],
    inplace=True
)
```

# Data Cleaning Summary

## Cleaning Actions Performed

1. Removed fully empty columns:
   - New
   - PendingS

2. Removed duplicate records.

3. Converted Date column to datetime format.

4. Investigated missing values in Amount and Currency.
   - Most missing values belong to cancelled orders.
   - These records were retained because they provide business insights.

5. Investigated fulfilled-by column.
   - Missing values appear to represent orders not fulfilled through Easy Ship.
   - Column retained for future analysis.

6. Removed rows with missing shipping information because they represented a very small portion of the dataset.

## Result

The dataset has been cleaned and prepared for exploratory data analysis (EDA).
