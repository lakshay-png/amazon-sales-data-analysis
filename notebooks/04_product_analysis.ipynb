#  Product Analysis

## Objective

The objective of this analysis is to identify the best-performing product categories and understand customer purchasing preferences.

Key Questions:
- Which categories generate the highest revenue?
- Which categories receive the most orders?
- Which product sizes are most popular?
- What business recommendations can be derived from product performance?

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/cleaned_amazon_sales.csv')
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df['Category'].value_counts()
```

```python
plt.figure(figsize=(12,6))

df['Category'].value_counts().plot(kind='bar')

plt.title('Orders by Category')
plt.xlabel('Category')
plt.ylabel('Number of Orders')

plt.xticks(rotation=45)

plt.show()
```

## Category Order Insights

- The most frequently ordered category is T-shirt.
- The least frequently ordered category is Watch.
- Customer demand is concentrated in T-shirt, Shirt and Blazzer categories.

```python
category_revenue = df.groupby('Category')['Amount'].sum().sort_values(ascending=False)

category_revenue
```

```python
plt.figure(figsize=(12,6))

category_revenue.plot(kind='bar')

plt.title('Revenue by Category')
plt.xlabel('Category')
plt.ylabel('Revenue')

plt.xticks(rotation=45)

plt.show()
```

## Revenue Insights

- Highest revenue-generating category: T-shirt
- Lowest revenue-generating category: Watch
- Revenue is heavily concentrated in: T-shirt,Shirt and Blazzer

```python
revenue_pct = (category_revenue / category_revenue.sum()) * 100

revenue_pct.round(2)
```

```python
revenue_pct.head()
```

```python
plt.figure(figsize=(8,8))

revenue_pct.head(5).plot(
    kind='pie',
    autopct='%1.1f%%'
)

plt.ylabel('')

plt.title('Top 5 Revenue Contributing Categories')

plt.show()
```

```python
df['Size'].value_counts()
```

```python
plt.figure(figsize=(12,6))

df['Size'].value_counts().plot(kind='bar')

plt.title('Size Distribution')
plt.xlabel('Size')
plt.ylabel('Orders')

plt.show()
```

## Size Insights

- Most popular size: M
- Least popular size: 4XL
- Majority of customers purchase: M,L and XL

```python
category_aov = df.groupby('Category')['Amount'].mean().sort_values(ascending=False)

category_aov
```

```python
plt.figure(figsize=(12,6))

category_aov.plot(kind='bar')

plt.title('Average Order Value by Category')
plt.xlabel('Category')
plt.ylabel('Average Order Value')

plt.xticks(rotation=45)

plt.show()
```

# Business Recommendations

1. Increase inventory for top-performing categories.
2. Prioritize marketing efforts toward high-revenue categories.
3. Monitor low-performing categories for optimization opportunities.
4. Focus production planning on the most popular product sizes.
5. Bundle complementary products to increase average order value.

# Key Findings

## Product Performance

1. Highest Revenue Category: T-shirt
2. Most Ordered Category: T-shirt
3. Most Popular Size: M
4. Highest Average Order Value Category: T-shirt

## Conclusion

Product demand is concentrated in a small number of categories. Strategic focus on top-performing categories can significantly improve revenue and operational efficiency.
