#Customer Segmentation Analysis

## Objective

The purpose of this analysis is to understand customer segments and compare the purchasing behavior of B2B and B2C customers.

Key Questions:

- What proportion of customers are B2B vs B2C?
- Which segment generates more revenue?
- Which segment places more orders?
- How does Average Order Value differ between segments?

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/content/cleaned_amazon_sales.csv')
```

```python
df['B2B'].value_counts()
```

```python
(df['B2B'].value_counts(normalize=True) * 100).round(2)
```

```python
plt.figure(figsize=(8,6))

df['B2B'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%'
)

plt.ylabel('')
plt.title('B2B vs B2C Orders')

plt.show()
```

## Customer Segment Insights

- B2C customers account for 99.3% of orders.
- B2B customers account for 0.7% of orders.
- The customer base is primarily B2C.

```python
segment_revenue = (
    df.groupby('B2B')['Amount']
    .sum()
)

segment_revenue
```

```python
plt.figure(figsize=(8,6))

segment_revenue.plot(kind='bar')

plt.title('Revenue by Customer Type')
plt.xlabel('B2B')
plt.ylabel('Revenue')

plt.show()
```

## Revenue Insights

- The highest revenue comes from B2C customers.
- Revenue contribution from B2B customers is 0.7%.

```python
df.groupby('B2B')['Order ID'].nunique()
```

```python
orders_by_segment = (
    df.groupby('B2B')['Order ID']
    .nunique()
)

orders_by_segment.plot(
    kind='bar',
    figsize=(8,6)
)

plt.title('Orders by Customer Type')

plt.show()
```

```python
segment_aov = (
    df.groupby('B2B')['Amount']
    .mean()
)

segment_aov
```

```python
segment_aov.plot(
    kind='bar',
    figsize=(8,6)
)

plt.title('Average Order Value by Customer Type')

plt.show()
```

```python
pd.crosstab(
    df['Category'],
    df['B2B']
)
```

```python
category_segment = pd.crosstab(
    df['Category'],
    df['B2B']
)

category_segment.plot(
    kind='bar',
    figsize=(14,6)
)

plt.title(
    'Category Preference by Customer Type'
)

plt.show()
```
