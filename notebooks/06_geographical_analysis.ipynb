# Geographical Analysis

## Objective

The purpose of this analysis is to understand geographical sales patterns and identify regions contributing the most orders and revenue.

Key Questions:

- Which states generate the most orders?
- Which states generate the highest revenue?
- Which cities contribute the most sales?
- How can business strategies be optimized regionally?

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/cleaned_amazon_sales.csv')
```

```python
state_orders = df['ship-state'].value_counts()

state_orders.head(10)
```

```python
plt.figure(figsize=(10, 6))
df['ship-state'].value_counts().head(10).plot(kind='bar')
plt.xlabel('State')
plt.ylabel('Number of Orders')
plt.xticks(rotation=45)
plt.title('Top 10 States with the Most Orders')
plt.show()
```

## State Order Insights

- State with the highest orders: Maharashtra
- State with the second-highest orders: Karnataka
- Customer demand is concentrated in: Maharashtra

```python
state_revenue = (
    df.groupby('ship-state')['Amount']
    .sum()
    .sort_values(ascending=False)
)

state_revenue.head(10)
```

```python
plt.figure(figsize=(12,6))

state_revenue.head(10).plot(kind='bar')

plt.title('Top 10 States by Revenue')
plt.xlabel('State')
plt.ylabel('Revenue')

plt.xticks(rotation=45)

plt.show()
```

## Revenue Insights

- Highest revenue generating state: Maharashtra
- Second highest revenue generating state: Karnataka
- Revenue is concentrated in: Maharashtra

```python
city_orders = df['ship-city'].value_counts()

city_orders.head(10)
```

```python
plt.figure(figsize=(12,6))

city_orders.head(10).plot(kind='bar')

plt.title('Top 10 Cities by Orders')
plt.xlabel('City')
plt.ylabel('Orders')

plt.xticks(rotation=45)

plt.show()
```

```python
city_revenue = (
    df.groupby('ship-city')['Amount']
    .sum()
    .sort_values(ascending=False)
)

city_revenue.head(10)
```

```python
plt.figure(figsize=(12,6))

city_revenue.head(10).plot(kind='bar')

plt.title('Top 10 Cities by Revenue')
plt.xlabel('City')
plt.ylabel('Revenue')

plt.xticks(rotation=45)

plt.show()
```

```python
state_pct = (
    state_revenue /
    state_revenue.sum()
) * 100

state_pct.head(10)
```

```python
plt.figure(figsize=(8,8))

state_pct.head(5).plot(
    kind='pie',
    autopct='%1.1f%%'
)

plt.ylabel('')
plt.title('Top 5 Revenue Contributing States')

plt.show()
```

# Business Insights

1. Focus marketing efforts on top-performing states.
2. Strengthen logistics infrastructure in high-demand regions.
